## 10.2 Aplicación de funciones: apply y map

```python
import pandas as pd
import numpy as np
```

### Ejercicio 1: Aplicar función lambda para elevar al cuadrado

```python
# Creamos una Serie numérica
serie = pd.Series([1, 2, 3, 4, 5])

# Aplicamos una función lambda para elevar al cuadrado cada valor
resultado = serie.apply(lambda x: x**2)

# Mostramos el resultado
resultado
```

### Ejercicio 2: Transformar nombres a mayúsculas usando map()

```python
# Creamos una Serie de nombres
nombres = pd.Series(["juan", "ana", "pedro", "maria"])

# Convertimos cada nombre a mayúsculas
resultado = nombres.map(lambda x: x.upper())

# Mostramos el resultado
resultado
```

### Ejercicio 3: Aplicar raíz cuadrada con apply()

```python
# Creamos una Serie numérica
numeros = pd.Series([4, 9, 16, 25, 36])

# Aplicamos la función de raíz cuadrada
resultado = numeros.apply(np.sqrt)

# Mostramos el resultado
resultado
```

### Ejercicio 4: Clasificar valores con función personalizada

```python
# Creamos una Serie de valores
valores = pd.Series([10, 45, 70, 30, 90])

# Creamos una función para clasificar los valores
def clasificar(valor):
    if valor >= 70:
        return "Alto"
    elif valor >= 40:
        return "Medio"
    else:
        return "Bajo"

# Aplicamos la función a la Serie
resultado = valores.apply(clasificar)

# Mostramos el resultado
resultado
```

### Ejercicio 5: Reemplazo usando map() con diccionario

```python
# Creamos una Serie de calificaciones
notas = pd.Series(["A", "B", "C", "A", "B"])

# Creamos un diccionario de reemplazo
equivalencias = {
    "A": "Excelente",
    "B": "Bueno",
    "C": "Regular"
}

# Aplicamos el reemplazo
resultado = notas.map(equivalencias)

# Mostramos el resultado
resultado
```

## 10.4 Filtrado y selección condicional

```python
# Creamos una Serie numérica
numeros = pd.Series([10, 55, 23, 87, 34, 91, 45])

# Filtramos los valores mayores que 50
resultado = numeros[numeros > 50]

# Mostramos el resultado
resultado
```

### Ejercicio 2: Mostrar los valores pares de una Serie

### Ejercicio 4: Usar una condición doble (>30 y <70)

## 10.7 Reindexación

### Ejercicio 1: Reindexar una Serie agregando un nuevo índice

```python
# Creamos una Serie con índices iniciales
serie = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])

# Reindexamos la Serie agregando un nuevo índice 'e'
serie_reindexada = serie.reindex(['a', 'b', 'c', 'd', 'e'])

# Mostramos el resultado
serie_reindexada
```

### Ejercicio 2: Reindexar una Serie en orden inverso

```python
# Creamos una Serie numérica
serie = pd.Series([5, 10, 15, 20], index=['a', 'b', 'c', 'd'])

# Reindexamos cambiando el orden del índice
serie_invertida = serie.reindex(['d', 'c', 'b', 'a'])

# Mostramos la Serie reordenada
serie_invertida
```

### Ejercicio 3: Reindexación usando fill_value

```python
# Creamos una Serie inicial
serie = pd.Series([100, 200, 300], index=['a', 'b', 'c'])

# Reindexamos agregando nuevos índices y llenando con 0
serie_reindexada = serie.reindex(['a', 'b', 'c', 'd', 'e'], fill_value=0)

# Mostramos la Serie resultante
serie_reindexada
```

### Ejercicio 4: Rellenar valores faltantes con el promedio

```python
# Creamos una Serie con algunos datos
serie = pd.Series([10, 20, 30], index=['a', 'b', 'c'])

# Reindexamos agregando nuevos índices
serie_reindexada = serie.reindex(['a', 'b', 'c', 'd', 'e'])

# Calculamos el promedio de los valores existentes
promedio = serie.mean()

# Llenamos los valores faltantes con el promedio
serie_reindexada = serie_reindexada.fillna(promedio)

# Mostramos el resultado
serie_reindexada
```

### Ejercicio 5: Reindexar con índices nuevos y existentes

```python
# Creamos una Serie inicial
serie = pd.Series([50, 60, 70], index=['x', 'y', 'z'])

# Definimos un nuevo índice que incluye valores existentes y nuevos
nuevo_indice = ['w', 'x', 'y', 'z', 'a']

# Aplicamos reindexación
serie_reindexada = serie.reindex(nuevo_indice)

# Mostramos el resultado
serie_reindexada
```

## Pregunta de reflexión

**¿Cómo contribuye el uso de Series en Pandas a la eficiencia y comprensión de los datos en un proyecto de análisis?**

### Respuesta

El uso de Series en Pandas mejora la eficiencia en el análisis de datos porque permite trabajar con datos de manera simple y organizada en una sola dimensión, similar a una columna de una tabla. Cada valor está asociado a un índice, lo que facilita localizar, seleccionar y modificar información específica sin necesidad de recorrer todo el conjunto de datos manualmente.

Además, las Series permiten aplicar operaciones de forma rápida sobre todos los elementos, como cálculos estadísticos, comparaciones o transformaciones. Esto agiliza el procesamiento de grandes volúmenes de datos y reduce el tiempo que el analista dedica a tareas repetitivas.

En un proyecto de análisis, las Series también ayudan a explorar y entender mejor los datos, ya que permiten detectar patrones, valores faltantes o comportamientos particulares dentro de una variable. De esta forma, se convierten en una herramienta clave para preparar la información antes de realizar análisis más complejos o construir modelos de datos.
